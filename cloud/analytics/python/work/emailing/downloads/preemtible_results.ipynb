```python
query = '''
SELECT
    billing_account_id,
    SUM(real_consumption) as paid,
    SUM(trial_consumption) as trial,
    SUM(real_consumption) + SUM(trial_consumption) as total

FROM
    cloud_analytics_testing.acquisition_cube_test
WHERE
    name LIKE '%preempt%'
    AND event = 'day_use'
    AND (real_consumption > 0 OR trial_consumption > 0)
    AND toDate(event_time) >= '2019-03-06'
GROUP BY
    billing_account_id
FORMAT TabSeparatedWithNames
'''

cloud_clickhouse_param_dict['query'] = query
cons = clickhouse.get_clickhouse_data(**cloud_clickhouse_param_dict)
```

```python
for col in ['paid','trial','total']:
    cons[col] = cons[col].astype(float)
    cons['convertion_' + col] = cons[col].apply(lambda x: 1 if x > 0 else 0)
    cons['users_' + col] = cons[col].apply(lambda x: 1 if x > 0 else 0)
    cons['convertion_10_' + col] = cons[col].apply(lambda x: 1 if x > 10 else 0)
    cons['usera_10_' + col] = cons[col].apply(lambda x: 1 if x > 10 else 0)
```

```python
res = pd.merge(
    sample,
    cons,
    on = 'billing_account_id',
    how = 'left'
).fillna(0)
```
